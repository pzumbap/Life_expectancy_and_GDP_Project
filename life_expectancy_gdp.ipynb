## Loading Data 

```python
import pandas as pd
from matplotlib import pyplot as plt
from numpy import mean
import seaborn as sns
```

```python
all_countries = pd.read_csv('all_data.csv')
#print(all_countries.head(), all_countries.info())
#Seems like the third column name are very descriptive but not usefull to use in programming.
all_countries.rename(columns={'Life expectancy at birth (years)': 'LEABY'}, inplace=True)
#print(all_countries.head(), all_countries.info())
```

## 1) Has life expectancy increased over time in the six nations?

## 2) Has GDP increased over time in the six nations?

```python
name_countries = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbawe']

ax1 = plt.subplots(figsize=(17, 5))
ax1 = plt.subplot(1, 2, 1)
ax1 = sns.set_style('whitegrid')
sns.lineplot(data=all_countries, x="Year", y="LEABY", hue='Country')
plt.title('Life Expectancy 2000-2015')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.legend(name_countries, loc=4)

ax2 = plt.subplot(1, 2, 2)
ax2 = sns.set_style('whitegrid')
sns.lineplot(data=all_countries, x="Year", y="GDP", hue='Country')
plt.title('Increase of GDP (Trillions of U.S. Dollars) | 2000-2015')
plt.xlabel('Years')
plt.ylabel('Billion Dollars')
plt.legend(name_countries, loc=2)

#plt.savefig('LEABY-GDP Lineplot.png')
plt.show()
```

In the first graph we can see that the country of Zimbawe has had a significant increase over the years with respect to life expectancy. The other countries seems to have a gradual small increase.

On the second graph with respect to GDP, we can see that USA and China are leading. Notice that US has had a small decrease on 2009 due the recession but inmmediately start to increase their GDP again. China has the most significant increase compared to the other countries. Germany has been slow fluctuating their GDP over the years. The other countries do not seem to have a significan increase over the years.

## 3) Is there a correlation between GDP and life expectancy of a country?

```python
plt.clf()

ax1 = plt.subplots(figsize=(10, 5))
ax1 = sns.set_style('whitegrid')
sns.scatterplot(data=all_countries, x="GDP", y="LEABY", hue='Country')
plt.title('Life expectancy vs GDP | 2000-2015')
plt.xlabel('GDP in Trillions of U.S. Dollars')
plt.ylabel('Life expectancy at birth (years)')

#plt.savefig('LEABY_vs_GDPscatterplot.png')
plt.show()
```

Here we have a scatterplot graph to represent the correlation between Life expectancy and GDP. Notice that Zimbawe only increase its Life expectancy without increase its GDP. Chile and Mexico have a small amount of increase on both instances, life expectancy and GDP. Germany increases its life expectancy but seems to be static in its GDP the last few years. USA and China have the same tendency to increase rapidly their GDP and slowly their life expectancy. We can notice again the USA recession of 2009 but this time we can see that happened when USA had around GDP=1.48 trillions of US Dollars.

## 4) What is the average life expactancy in these nations?

```python
plt.clf()

ax1 = plt.subplots(figsize=(10, 5))
ax1 = sns.set_style('whitegrid')
sns.barplot(data=all_countries, x="Country", y="LEABY", estimator=mean, linewidth=2, edgecolor=".2")
plt.xticks(rotation=5)
plt.title('Average life expactancy | 2000-2015')
plt.xlabel('Country')
plt.ylabel('Average life expectancy at birth (years)')

#plt.savefig('AverageLifeExpectancy.png')
plt.show()
```

In this graph we can see cleary that the average life expectancy of Chile, China, Germany, Mexico and USA are between the age of 73 and 80. Only Zimbawe is far behind from them with an average life expectancy of 50 years old. Notice that the error bar of Zimbawe is big compared to the other countries; this is because the high variation of life expectancy over the years.

## 5) What is the distribution of that life expectancy?

```python
plt.clf()

ax1 = plt.subplots(figsize=(15, 7))
ax1 = sns.set_style('darkgrid')
sns.boxplot(data=all_countries, x="Country", y="LEABY", linewidth=1)
sns.swarmplot(data=all_countries, x="Country", y="LEABY", color="0.25")
plt.title('Distribution of life expectancy by country | 2000-2015')
plt.xlabel('Country')
plt.ylabel('Average life expectancy at birth (years)')

#plt.savefig('AverageLifeExpectancy.png')
plt.show()
```

Here we can see in more detail the distribution of life expectancy in each country. Chile and Germany lead in life expectancy despite having a lower GDP than the United States and China. Mexico has a very narrow distribution, so its life expectancy is almost the same over the years. On the other hand, Zimbabwe has a significant distribution that explains the length of the error bar in the previous graph.

# Conclusions

- All countries are gradually increasing their life expectancy, but Zimbabwe seems to be increasing faster.
- The United States is the country that leads the GDP growth, but China's GDP is growing very fast.
- It seems that GDP directly influences life expectancy, but it is not very clear when we look at Zimbabwe as it does not show an increase in its GDP. However, this does not mean that Zimbabwe's GDP is static, it could be that Zimbabwe's GDP is increasing, but on a different monetary scale; as the data is presented in trillions of US dollars.
- Average life expectancy is not a very descriptive measure. To get a better understanding, we can use a boxplot. There we can see that there are no outliers for each country and most of the countries have a narrow dispersion, except Zimbabwe.
